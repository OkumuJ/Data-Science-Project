## Empty Cells
#### Remove Rows
new_df = df.dropna() - this returns new data frame without empty cells
the dropna() method returns a new DataFrame, and will not change the original df. If changes are to be implemented on the intial data frame, we use the inplace=True

#### Replace empty Cells 
fillna() method allows us to replace empty cells with a value 
> df.fillna(123, inplace=True)

#### Replace only specified cells
>df['columns_name'].fillna(130, inplace=True)


### Replace using Mean, Median, or Mode
example
> x = df['columns_name'].mean()
> df['columns_name'].fillna(x, inplace=True)

> x = df['columns_name'].median()
> df['columns_name'].fillna(x, inplace=True)

> x = df['columns_name'].mode()
> df['columns_name'].fillna(x, inplace=True)
